#ข้อมูลชุดที่ 3

```python
path = '/content/drive/My Drive/DATA' #path คือที่อยู่ของไฟล์
data3  = pd.read_csv(os.path.join(path,'10-2564-1.csv'))
data3
```

# ความเชื่อมโยงของ 3 ตาราง (Index)

## เชื่อม 2 ตารางก่อน

```python
data2_new = data2.iloc[:,[4,6,7]] #ตัดข้อมูลด้วย iloc เอาแค่คอลัมม์ 	Level_School ,	WeightHigh	และ Total WeightHigh
data2_new
```

##การรวมข้อมูล

*   1.  รวม 2 ตารางเข้าด้วยกัน
*   2.  เลือกมาเฉพาะบาง column จากอีกตารางหนึ่งมาเเปะไว้



## รวม 2 ตาราง (.merge())

```python
merged_table1 = data1.merge(data2_new,how='left',left_on='Level_School',right_on='Level_School') #รวมตาราง ซึ่งเอาข้อมูลชุดที่ 1 ไว้ทางด้านซ้าย ข้อมูลชุดที่ 2 ไว้ทางขวา และข้อมูลชุดที่ 1 เป็นตารางหลัก เชื่อมกันด้วย Level_School 
merged_table1
```

## เชื่อมโยงกัน  3 ตาราง 

```python
data3_new = data3.iloc[:,[4,6,7]] #ตัดข้อมูลด้วย iloc เอาแค่คอลัมม์  School_ID, Nationality และ Total_Nationality
data3_new
```

```python
merged_table_new = merged_table1.merge(data3_new,how='left',left_on='Level_School',right_on='Level_School') #ตารางเชื่อมโยงกัน 3 ตาราง ซึ่งเอาข้อมูลที่อยู่ทางซ้ายเป็นข้อมูลที่เชื่อมกัน 2 ตาราง  ส่วนตาราง data3_new ไว้ทางขวา ตารางหลักคือตารางด้านซ้ายเชื่อมกันด้วย Level_School 
merged_table_new
```
