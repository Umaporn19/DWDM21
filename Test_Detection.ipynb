```python
#Numpy and Image Assessment

import matplotlib.pyplot as plt
import urllib.request
import numpy as np

%matplotlib inline
from PIL import Image


urllib.request.urlretrieve(
  'https://static.wikia.nocookie.net/ladygaga/images/9/9a/Blackpink.jpg',
   "bkp.png")
  
v_pic = Image.open("bkp.png")

#v_pic
```

```python
v_pic_arr = np.asarray(v_pic)
v_pic_arr.shape #pixel width * pixel height * color channels 
```

```python
# Show all
v_pic_arr[:,:,:]
```

```python
# Use PIL and Matplotlib to read and display 
plt.imshow(v_pic_arr)
```

```python
# R G B
plt.imshow(v_pic_arr[:,:,0],cmap = 'gray') #Red Channel Values 0-255 // 0 = no color
```

```python
# R G B
plt.imshow(v_pic_arr[:,:,1],cmap = 'gray') #Green Channel
```

```python
# R G B
plt.imshow(v_pic_arr[:,:,2],cmap = 'gray') #Blue Channel
```

```python
#Copy data
pic_blue = v_pic_arr.copy()
```

```python
#Use slicing to set the RED and GREEN channels of the picture to 0, then use imshow() to show the isolated blue channel
pic_blue[:, :, 0] = 0    # Zero out contribution from red
pic_blue[:, :, 1] = 0   # Zero out contribution from green
plt.imshow(pic_blue)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2

blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)

blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
# pt1 = top left
# pt2 = bottom right
cv2.rectangle(blank_img,pt1=(400,0),pt2=(500,200),color=(0,255,0),thickness=5)
plt.imshow(blank_img)
```

```python
#Drawing a blue rectangle in the middle of the image.
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(blank_img)
```
